```python
whodf = whodf.sample(180)
```

```python
X = whodf[list(set(whodf.columns)-set(["Gross national income per capita (PPP international $)", "Continent"]))].values
y = whodf["Gross national income per capita (PPP international $)"].values

print (X.shape, y.shape)
```

```python
from sklearn.manifold import TSNE
df = TSNE(verbose=True)
```

```python
X_df = df.fit_transform(X)
```

#### For visualizing with t-SNE I willuse ggplot. First I install from the prompt with "pip install -U ggplot" comand. Then I load the package here

```python
from ggplot import *
```

```python
df_tsne = whodf.loc[:,:].copy()
df_tsne['x-tsne'] = X_df[:,0]
df_tsne['y-tsne'] = X_df[:,1]

dfplot = ggplot( df_tsne, aes(x='x-tsne', y='y-tsne', color='Gross national income per capita (PPP international $)')) \
        + geom_point(size=150,alpha=0.8) \
        + ggtitle("tSNE dimensions colored by GNIPPP")
dfplot
```
