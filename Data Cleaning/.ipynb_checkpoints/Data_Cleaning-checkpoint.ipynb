# The aim of this notebook is to clean WHO dataset for further analysis.

#### First I load some useful packagues and WHO dataset from DATASET directory (You can also download to the follow website. (https://data.world/resiport/who-dataset)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
who = pd.read_csv("../DATASETS/WHO.csv", sep = ',', decimal = '.' ) 
```

```python
who
```

#### I change row names for "Country" column and I put in "Continent" his name.

```python
who = who.set_index("Country")
```

```python
vals_to_replace = {1: 'WestAsia', 2: 'Europe', 3:'Africa', 4:'AmericaN', 5:'AmericaS',
                                         6:'SouthAsia-Oceania', 7:'EastAsia'}
who['Continent']= who['Continent'].map(vals_to_replace)
```

```python
who.head(5)
```

#### I delete columns which have more than 50% NAs (it is 100 NAs), because it's not present enough imformation.

```python
numbernas = who.isnull().apply(sum, axis=0) # count the number of nan in each column
print(numbernas)
```

```python
for col in who: 
   if numbernas[col] >= 100:  
       del who[col]
```

```python
list(who)
```

#### I'll try to impute the median by columns for NaN values depending on the country to which it corresponds. 
#### First of all I'm going to split the dataset by continent, second I impute the median and check that there isn't NA's.
#### Finally I merge the resulting datasets.

```python
westasia = who[who.Continent == "WestAsia"]
westasiaC = westasia.fillna(westasia.median(), inplace=True)
westasiaC
```

```python
westasiaC.isnull().values.any() # Check if there are any NA's values.
```

```python
europe = who[who.Continent == "Europe"]
europeC = europe.fillna(europe.median(), inplace=True)
europeC
```

```python
europeC.isnull().values.any()
```

```python
africa = who[who.Continent == "Africa"]
africaC = africa.fillna(africa.median(), inplace=True)
africaC   
```

```python
africaC.isnull().values.any()
```

```python
americaN = who[who.Continent == "AmericaN"]
americaNC = americaN.fillna(americaN.median(), inplace=True)
americaNC 
```

```python
americaNC.isnull().values.any()
```

```python
americaS = who[who.Continent == "AmericaS"]
americaSC = americaS.fillna(americaS.median(), inplace=True)
americaSC
```

```python
americaSC.isnull().values.any()
```

```python
sasiaoceania = who[who.Continent == "SouthAsia-Oceania"]
sasiaoceaniaC = sasiaoceania.fillna(sasiaoceania.median(), inplace=True)
sasiaoceaniaC
```

```python
sasiaoceaniaC.isnull().values.any()
```

```python
eastasia = who[who.Continent == "EastAsia"]
eastasiaC = eastasia.fillna(eastasia.median(), inplace=True)
eastasiaC
```

```python
eastasiaC.isnull().values.any() # Here we have any missing value. Let's look for it!
```

```python
eastasiaC.isnull().any() 
```

#### We have missing values in "One-year-olds immunized with three doses of Hib (Hib3) vaccine (%)" because there won't be data for any country in thisfeature, so I drop this column when I merge all dataset.

```python
# Here I merge the dataset by continent
whocomplete1 = westasiaC.append(europeC)
whocomplete2 = whocomplete1.append(africaC)
whocomplete3 = whocomplete2.append(americaNC)
whocomplete4 = whocomplete3.append(americaSC)
whocomplete5 = whocomplete4.append(sasiaoceaniaC)
whocomplete = whocomplete5.append(eastasiaC)

whocomplete
```

```python
whocomplete = whocomplete.sort_values(by= "CountryID", ascending=True) #Order by countryID the dataset, like the first one.
whocomplete
```

```python
whocomplete= whocomplete.drop("CountryID", 1) # Delete "CountryID" column.
```

```python
whocomplete= whocomplete.drop("One-year-olds immunized with three doses of Hib (Hib3) vaccine (%)", 1) 
# Delete "One-year-olds immunized with three doses of Hib (Hib3) vaccine (%)" column.
```

```python
whocomplete
```

#### Last I export the dataset and move to DATASETS directory

```python
# Now I export "WHOCOMPLETE" dataset in CSV format for further analysis
whocomplete.to_csv("whocomplete.csv", sep = ";", decimal= ".")
```
