# The aim of this notebook is to clean WHO dataset for further analysis.

#### First I load some useful packagues and WHO dataset to the follow website. (https://data.world/resiport/who-dataset)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
who = pd.read_csv('https://query.data.world/s/8Pb1O_ASzuDGfaFfGu6w6kLT6t6IwL', sep = ',', decimal = '.' )
```

```python
who
```

#### I change row names for "Country" column and I put in "Continent" his name.

```python
who = who.set_index("Country")
```

```python
vals_to_replace = {1: 'WestAsia', 2: 'Europe', 3:'Africa', 4:'AmericaN', 5:'AmericaS',
                                         6:'SouthAsia-Oceania', 7:'EastAsia'}
who['Continent']= who['Continent'].map(vals_to_replace)
```

```python
who.head(5)
```

#### I delete columns which have more than 50% NAs (it is 100 NAs), because it's not present enough imformation.

```python
numbernas = who.isnull().apply(sum, axis=0) # count the number of nan in each column
print(numbernas)
```

```python
for col in who: 
   if numbernas[col] >= 100:  
       del who[col]
```

```python
list(who)
```

#### I'll try to impute the median by columns for NaN values depending on the country to which it corresponds. 
#### First of all I'm going to split the dataset by continent, second I impute the median and finally I merge the resulting datasets.

```python
westasia = who[who.Continent == "WestAsia"]
westasiaC = westasia.fillna(westasia.median(), inplace=True)
westasiaC.head(5) 
```

```python
europe = who[who.Continent == "Europe"]
europeC = europe.fillna(europe.median(), inplace=True)
europeC.head(5) 
```

```python
africa = who[who.Continent == "Africa"]
africaC = africa.fillna(africa.median(), inplace=True)
africaC.head(5)    
```

```python
americaN = who[who.Continent == "AmericaN"]
americaNC = americaN.fillna(americaN.median(), inplace=True)
americaNC.head(5) 
```

```python
americaS = who[who.Continent == "AmericaS"]
americaSC = americaS.fillna(americaS.median(), inplace=True)
americaSC.head(5)
```

```python
sasiaoceania = who[who.Continent == "SouthAsia-Oceania"]
sasiaoceaniaC = sasiaoceania.fillna(sasiaoceania.median(), inplace=True)
sasiaoceaniaC.head(5)
```

```python
eastasia = who[who.Continent == "EastAsia"]
eastasiaC = eastasia.fillna(eastasia.median(), inplace=True)
eastasiaC.head(5)
```

```python
whocomplete1 = westasiaC.append(europeC)
whocomplete2 = whocomplete1.append(africaC)
whocomplete3 = whocomplete2.append(americaNC)
whocomplete4 = whocomplete3.append(americaSC)
whocomplete5 = whocomplete4.append(sasiaoceaniaC)
whocomplete = whocomplete5.append(eastasiaC)

whocomplete.head(2)
```

```python
whocomplete = whocomplete.sort_values(by= "CountryID", ascending=True)
whocomplete
```
